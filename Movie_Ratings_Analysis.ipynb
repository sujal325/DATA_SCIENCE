```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('/content/links.csv')
df2 = pd.read_csv('/content/movies.csv')
df3 = pd.read_csv('/content/ratings.csv')
df4 = pd.read_csv('/content/tags.csv')


data = df1.merge(df2,on='movieId').merge(df3,on='movieId').merge(df4,on='movieId')
data.head()
```

```python
average_ratings = data.groupby('tag').rating.mean().reset_index()
average_ratings.head()
```

```python
top20_tags = average_ratings.nlargest(20,'rating')
print(top20_tags)
```

```python
sns.barplot(x = 'rating',y = 'tag',data= top20_tags,palette='viridis')
plt.xlabel('highest ratin5')
plt.ylabel('production house')
plt.title('Top 20 higest rated movies production house')
```
